# **1. Perkenalan Dataset**

## 1.1 Tentang Dataset

Dataset yang digunakan dalam analisis ini berjudul "Bank Transaction Dataset for Fraud Detection" yang tersedia di platform Kaggle. Dataset ini dapat diakses melalui tautan berikut:

[https://www.kaggle.com/datasets/valakhorasani/bank-transaction-dataset-for-fraud-detection](https://www.kaggle.com/datasets/valakhorasani/bank-transaction-dataset-for-fraud-detection)

Dataset ini diunggah oleh penyumbang bernama Val Akhorasani, dan dirancang khusus untuk analisis dan pengembangan model deteksi penipuan (fraud detection) dalam sistem perbankan. Dataset ini mencakup berbagai atribut komprehensif yang merepresentasikan transaksi perbankan dari berbagai aspek, termasuk informasi transaksi, detail nasabah, dan metadata teknis.

## 1.2 Struktur Dataset

Dataset ini memiliki struktur yang kaya dengan 16 kolom yang dapat dikelompokkan ke dalam beberapa kategori utama:

### 1.2.1 Identifikasi Transaksi
- **TransactionID**: Pengidentifikasi unik untuk setiap transaksi yang memungkinkan pelacakan individu setiap transaksi dalam sistem.
- **AccountID**: Pengidentifikasi unik untuk akun nasabah, memungkinkan analisis perilaku nasabah secara konsisten.

### 1.2.2 Informasi Transaksi
- **TransactionAmount**: Jumlah uang yang ditransaksikan, menunjukkan besaran nilai transaksi.
- **TransactionDate**: Timestamp yang menunjukkan kapan transaksi dilakukan, memungkinkan analisis temporal.
- **TransactionType**: Jenis transaksi (Debit/Credit), mengindikasikan arah aliran dana.
- **PreviousTransactionDate**: Tanggal dan waktu transaksi sebelumnya, memungkinkan analisis frekuensi transaksi.
- **TransactionDuration**: Durasi transaksi dalam detik, dapat mengindikasikan kompleksitas atau anomali transaksi.

### 1.2.3 Informasi Lokasi dan Saluran
- **Location**: Lokasi geografis di mana transaksi dilakukan, penting untuk analisis spasial.
- **Channel**: Saluran atau metode yang digunakan untuk transaksi (misalnya ATM, Online, Mobile, Branch), menunjukkan preferensi nasabah.
- **MerchantID**: Pengidentifikasi pedagang atau penerima pembayaran, memungkinkan analisis pola pembelanjaan.

### 1.2.4 Informasi Teknis dan Keamanan
- **DeviceID**: Pengidentifikasi perangkat yang digunakan untuk transaksi, penting untuk deteksi akses tidak sah.
- **IP Address**: Alamat IP yang digunakan saat transaksi, dapat mengindikasikan lokasi digital dan potensi anomali.
- **LoginAttempts**: Jumlah percobaan login sebelum transaksi berhasil, indikator potensial usaha peretasan.

### 1.2.5 Informasi Nasabah
- **CustomerAge**: Usia nasabah, memberikan informasi demografis penting.
- **CustomerOccupation**: Pekerjaan nasabah, dapat membantu memahami segmentasi pelanggan.
- **AccountBalance**: Saldo akun setelah transaksi, memberikan konteks finansial untuk analisis.

Dataset transaksi perbankan dari Kaggle ini memiliki beberapa aspek signifikan yang membuatnya bernilai untuk analisis clustering dan deteksi penipuan:

Dataset bank transaction dari Val Akhorasani ini membuka berbagai kemungkinan analisis, terutama dengan pendekatan clustering:

# **5. Data Preprocessing**

Berdasarkan hasil EDA sebelumnya, saya akan membangun solusi clustering untuk data transaksi bank dengan fokus pada deteksi pola-pola yang dapat membantu identifikasi penipuan. Mari kita mulai dengan preprocessing data yang komprehensif dan dilanjutkan dengan pembangunan model clustering yang optimal.

## **5.1 Menghapus Kolom yang Tidak Relevan**
Untuk proses clustering, kita akan menghapus kolom-kolom identifikasi yang tidak relevan dengan pola transaksi:

```python
# Menghapus kolom identifikasi yang tidak diperlukan untuk clustering
id_columns = ['TransactionID', 'AccountID', 'DeviceID', 'IP Address', 'MerchantID']
date_columns = ['TransactionDate', 'PreviousTransactionDate']
df_clean = df.drop(columns=id_columns + date_columns)
print("Dimensi dataset setelah menghapus kolom ID:", df_clean.shape)
```

## **5.2 Konversi Data Kategorikal**
Algoritma clustering membutuhkan data numerik, jadi kita perlu mengkonversi data kategorikal:

```python
# Konversi TransactionType menggunakan binary encoding
df_clean['TransactionType'] = df_clean['TransactionType'].map({'Credit': 1, 'Debit': 0})

# Konversi Channel menggunakan one-hot encoding
channel_dummies = pd.get_dummies(df_clean['Channel'], prefix='Channel', drop_first=False)
df_clean = pd.concat([df_clean.drop(columns=['Channel']), channel_dummies], axis=1)

# Konversi CustomerOccupation menggunakan one-hot encoding
occupation_dummies = pd.get_dummies(df_clean['CustomerOccupation'], prefix='Occupation', drop_first=False)
df_clean = pd.concat([df_clean.drop(columns=['CustomerOccupation']), occupation_dummies], axis=1)

# Konversi Location menggunakan frequency encoding untuk menangani banyaknya kategori
location_freq = df_clean['Location'].value_counts(normalize=True).to_dict()
df_clean['Location_Freq'] = df_clean['Location'].map(location_freq)
df_clean = df_clean.drop(columns=['Location'])
```

## **5.3 Feature Engineering**
Kita akan menciptakan fitur baru yang dapat meningkatkan kualitas clustering:

```python
# Ekstrak fitur waktu dari TransactionDate
df_clean['TransactionHour'] = pd.to_datetime(df['TransactionDate']).dt.hour
df_clean['TransactionDayOfWeek'] = pd.to_datetime(df['TransactionDate']).dt.dayofweek

# Tambahkan fitur yang menunjukkan waktu dari transaksi terakhir (dalam jam)
df_clean['TimeSincePreviousTransaction'] = (pd.to_datetime(df['TransactionDate']) -
                                         pd.to_datetime(df['PreviousTransactionDate'])).dt.total_seconds() / 3600

# Fitur terkait perilaku transaksi
df_clean['TransactionAmountToBalance'] = df_clean['TransactionAmount'] / (df_clean['AccountBalance'] + 1)  # Rasio jumlah transaksi terhadap saldo
df_clean['LoginToTransactionRatio'] = df_clean['LoginAttempts'] / (df_clean['TransactionDuration'] + 1)  # Rasio login terhadap durasi

# Fitur pola waktu
df_clean['Is_Weekend'] = (df_clean['TransactionDayOfWeek'] >= 5).astype(int)
df_clean['Is_BusinessHours'] = ((df_clean['TransactionHour'] >= 9) & (df_clean['TransactionHour'] <= 17)).astype(int)
df_clean['Is_LateNight'] = ((df_clean['TransactionHour'] >= 22) | (df_clean['TransactionHour'] <= 5)).astype(int)
```

## **5.4 Deteksi dan Penanganan Outlier**
Outlier dapat mempengaruhi kualitas clustering, jadi kita akan mendeteksi dan menanganinya:

```python
# Fungsi untuk mendeteksi outlier menggunakan Isolation Forest
def handle_outliers(df, contamination=0.05):
    # Pilih kolom numerik untuk deteksi outlier
    num_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()

    # Terapkan Isolation Forest
    iso_forest = IsolationForest(contamination=contamination, random_state=42)
    outlier_labels = iso_forest.fit_predict(df[num_cols])

    # Konversi ke indikator biner (1: normal, -1: outlier)
    outlier_indicators = pd.Series(outlier_labels, index=df.index)
    outlier_count = (outlier_indicators == -1).sum()

    print(f"Isolation Forest menemukan {outlier_count} outlier ({outlier_count/len(df)*100:.2f}%)")

    # Buat kolom bobot (outlier mendapat bobot lebih rendah)
    df['sample_weight'] = 1.0
    df.loc[outlier_indicators == -1, 'sample_weight'] = 0.2

    return df

# Terapkan deteksi outlier
df_clean = handle_outliers(df_clean)
```

# **6. Pembangunan Model Clustering**

## **a. Pembangunan Model Clustering**

Pada tahap ini, Anda membangun model clustering dengan memilih algoritma yang sesuai untuk mengelompokkan data berdasarkan kesamaan.

```python
# Siapkan data untuk clustering
X = df_clean.drop(columns=['sample_weight']).values
weights = df_clean['sample_weight'].values

# Inisialisasi model KMeans
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)

# Latih model dengan mempertimbangkan bobot sample
kmeans.fit(X, sample_weight=weights)

# Simpan hasil clustering
df_clean['cluster'] = kmeans.labels_
```

## **b. Evaluasi Model Clustering**

Untuk menentukan jumlah cluster yang optimal dalam model clustering, Anda dapat menggunakan metode Elbow atau Silhouette Score.

Metode ini membantu kita menemukan jumlah cluster yang memberikan pemisahan terbaik antar kelompok data, sehingga model yang dibangun dapat lebih efektif. Berikut adalah **rekomendasi** tahapannya.
1. Gunakan Silhouette Score dan Elbow Method untuk menentukan jumlah cluster optimal.
2. Hitung Silhouette Score sebagai ukuran kualitas cluster.

```python
# Fungsi untuk menghitung skor untuk berbagai jumlah cluster
def evaluate_kmeans(X, weights, max_clusters=10):
    inertia_values = []
    silhouette_scores = []

    for n_clusters in range(2, max_clusters + 1):
        # Latih model KMeans
        kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
        kmeans.fit(X, sample_weight=weights)

        # Simpan inertia (metode elbow)
        inertia_values.append(kmeans.inertia_)

        # Hitung Silhouette Score
        cluster_labels = kmeans.labels_
        silhouette_avg = silhouette_score(X, cluster_labels)
        silhouette_scores.append(silhouette_avg)

        print(f"Untuk {n_clusters} cluster, silhouette score: {silhouette_avg:.4f}")

    return inertia_values, silhouette_scores

# Evaluasi model dengan berbagai jumlah cluster
inertia_values, silhouette_scores = evaluate_kmeans(X, weights)

# Plot Elbow Method
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(range(2, 11), inertia_values, 'bo-')
plt.title('Metode Elbow')
plt.xlabel('Jumlah Cluster')
plt.ylabel('Inertia')

# Plot Silhouette Score
plt.subplot(1, 2, 2)
plt.plot(range(2, 11), silhouette_scores, 'ro-')
plt.title('Silhouette Score')
plt.xlabel('Jumlah Cluster')
plt.ylabel('Silhouette Score')
plt.tight_layout()
plt.show()

# Identifikasi jumlah cluster optimal
optimal_clusters = silhouette_scores.index(max(silhouette_scores)) + 2
print(f"Jumlah cluster optimal berdasarkan Silhouette Score: {optimal_clusters}")
print(f"Silhouette Score tertinggi: {max(silhouette_scores):.4f}")
```

## **c. Feature Selection (Opsional)**

Silakan lakukan feature selection jika Anda membutuhkan optimasi model clustering. Jika Anda menerapkan proses ini, silakan lakukan pemodelan dan evaluasi kembali menggunakan kolom-kolom hasil feature selection. Terakhir, bandingkan hasil performa model sebelum dan sesudah menerapkan feature selection.

```python
# Feature Selection untuk optimasi model clustering
from sklearn.feature_selection import SelectKBest, mutual_info_regression, f_classif
```

```python
# Feature selection menggunakan metode SelectKBest dengan mutual_info_regression
def select_best_features(X, n_features=5):
    # Buat variabel target artifisial berdasarkan jarak ke centroid untuk supervised selection
    kmeans_temp = KMeans(n_clusters=3, random_state=42, n_init=10)
    kmeans_temp.fit(X)
    distances = kmeans_temp.transform(X).min(axis=1)

    # Gunakan Mutual Information untuk memilih fitur terbaik
    selector = SelectKBest(mutual_info_regression, k=n_features)
    X_new = selector.fit_transform(X, distances)

    # Dapatkan indeks fitur yang dipilih
    selected_indices = selector.get_support(indices=True)

    return X_new, selected_indices

# Pilih 5 fitur terbaik
X_selected, selected_indices = select_best_features(X, n_features=5)
selected_features = df_clean.drop(columns=['sample_weight', 'cluster']).columns[selected_indices].tolist()

print(f"Fitur yang dipilih: {selected_features}")
```

```python
# Evaluasi model dengan fitur terpilih
inertia_selected, silhouette_selected = evaluate_kmeans(X_selected, weights)

# Plot perbandingan Silhouette Score
plt.figure(figsize=(10, 6))
plt.plot(range(2, 11), silhouette_scores, 'bo-', label='Semua Fitur')
plt.plot(range(2, 11), silhouette_selected, 'ro-', label='Fitur Terpilih')
plt.title('Perbandingan Silhouette Score')
plt.xlabel('Jumlah Cluster')
plt.ylabel('Silhouette Score')
plt.legend()
plt.grid(True)
plt.show()

# Latih model final dengan fitur terpilih dan jumlah cluster optimal
optimal_clusters_selected = silhouette_selected.index(max(silhouette_selected)) + 2
kmeans_selected = KMeans(n_clusters=optimal_clusters_selected, random_state=42, n_init=10)
kmeans_selected.fit(X_selected, sample_weight=weights)

print(f"Silhouette Score tertinggi dengan semua fitur: {max(silhouette_scores):.4f}")
print(f"Silhouette Score tertinggi dengan fitur terpilih: {max(silhouette_selected):.4f}")

# Tambahkan hasil clustering dengan fitur terpilih ke dataframe
df_clean['cluster_selected'] = kmeans_selected.labels_
```

**Mengembalikan Data ke Skala Asli**

Setelah melakukan normalisasi atau standardisasi, kita perlu mengembalikan data ke skala aslinya agar lebih mudah diinterpretasikan. Proses ini melibatkan beberapa langkah:
- Membaca Data Asli - Kita memuat dataset sebelum dilakukan scaling untuk mendapatkan nilai awalnya.
- Identifikasi Kolom yang Ditransformasi - Menentukan fitur numerik yang telah diubah selama proses normalisasi.
- Inverse Transform - Menggunakan RobustScaler untuk mengembalikan setiap fitur yang telah dinormalisasi ke bentuk aslinya.

```python
# Kita perlu memuat dataset asli
df_raw = df.copy()

# Mendapatkan kolom yang telah discaling
numerical_cols = df_raw.select_dtypes(include=['int64', 'float64']).columns.tolist()
if 'sample_weight' in numerical_cols:
    numerical_cols.remove('sample_weight')
scaled_features = [col for col in df_original.columns if col in numerical_cols and col != 'Cluster' and col != 'sample_weight']

# Melakukan inverse transform pada fitur yang discaling
df_inverted = df_original.copy()
for col in scaled_features:
    if col in df_original.columns:
        # Membuat scaler khusus untuk kolom ini
        single_col_scaler = RobustScaler()
        # Fit scaler hanya dengan kolom ini dari data asli
        single_col_scaler.fit(df_raw[[col]])

        # Reshape ke format yang dibutuhkan oleh inverse_transform
        scaled_values = df_original[col].values.reshape(-1, 1)
        original_values = single_col_scaler.inverse_transform(scaled_values)
        df_inverted[col] = original_values.flatten()
```

**Persiapan Analisis Karakteristik Cluster**

Sebelum melakukan analisis lebih lanjut, kita perlu menentukan fitur-fitur yang akan digunakan untuk memahami karakteristik tiap cluster:

```python
# Konversi fitur boolean ke tipe kategorikal/boolean
boolean_features = ['Is_Late_Night', 'Is_Weekend', 'Is_BusinessHours']
for feature in boolean_features:
    if feature in df_inverted.columns:
        df_inverted[feature] = df_inverted[feature].astype('bool')

# Identifikasi fitur-fitur yang ada di df_inverted
numerical_features = df_inverted.select_dtypes(include=['int64', 'float64']).columns.tolist()
# Hapus 'Cluster' dari daftar fitur numerik jika ada
if 'Cluster' in numerical_features:
    numerical_features.remove('Cluster')
if 'sample_weight' in numerical_features:
    numerical_features.remove('sample_weight')
# Identifikasi fitur kategorikal di df_inverted
categorical_features = df_inverted.select_dtypes(include=['object', 'bool', 'category']).columns.tolist()
```

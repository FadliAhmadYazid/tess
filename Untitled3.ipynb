### **4.2 Fungsi Preprocessing**

```python
def preprocess_text(text):
    """
    Fungsi untuk melakukan preprocessing pada teks ulasan
    """
    if not isinstance(text, str):
        return ""

    # 1. Lowercasing
    text = text.lower()

    # 2. Hapus URL
    text = re.sub(r'https?://\S+|www\.\S+', '', text)

    # 3. Hapus emoji dan emoticon
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F700-\U0001F77F"  # alchemical symbols
                               u"\U0001F780-\U0001F7FF"  # Geometric Shapes
                               u"\U0001F800-\U0001F8FF"  # Supplemental Arrows-C
                               u"\U0001F900-\U0001F9FF"  # Supplemental Symbols and Pictographs
                               u"\U0001FA00-\U0001FA6F"  # Chess Symbols
                               u"\U0001FA70-\U0001FAFF"  # Symbols and Pictographs Extended-A
                               u"\U00002702-\U000027B0"  # Dingbats
                               u"\U000024C2-\U0001F251"
                               "]+", flags=re.UNICODE)
    text = emoji_pattern.sub(r'', text)

    # 4. Hapus punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))

    # 5. Hapus angka
    text = re.sub(r'\d+', '', text)

    # 6. Hapus karakter berulang
    text = re.sub(r'(.)\1+', r'\1\1', text)  # Reduksi karakter berulang lebih dari 2 menjadi 2

    # 7. Tokenisasi
    tokens = word_tokenize(text)

    # 8. Normalisasi kata tidak baku
    normalized_tokens = []
    for token in tokens:
        normalized_token = normalization_dict.get(token, token)
        if normalized_token:  # Skip jika token dinormalisasi menjadi string kosong
            normalized_tokens.append(normalized_token)

    # 9. Hapus stopwords
    stop_words = set(stopwords.words('indonesian'))
    additional_stopwords = {'nya', 'yg', 'di', 'dgn', 'krn', 'dg', 'tp', 'tapi', 'ke', 'ga', 'gak', 'jg', 'juga', 'dan', 'atau', 'dan', 'yang', 'dari', 'untuk', 'ke', 'pada'}
    stop_words.update(additional_stopwords)
    filtered_tokens = [word for word in normalized_tokens if word not in stop_words]

    # 10. Stemming untuk bahasa Indonesia menggunakan Sastrawi
    factory = StemmerFactory()
    stemmer = factory.create_stemmer()
    stemmed_tokens = [stemmer.stem(token) for token in filtered_tokens]

    # 11. Filter kata pendek (kurang dari 3 huruf)
    filtered_tokens = [token for token in stemmed_tokens if len(token) >= 3]

    # 12. Gabungkan token menjadi teks
    processed_text = ' '.join(filtered_tokens)

    return processed_text
```

### **5.3 Top N-gram Setelah Preprocessing**

```python
def plot_top_ngrams(df, sentiment, n=1, top=20):
    texts = df[df['sentiment'] == sentiment]['processed_text'].tolist()
    words = []
    for text in texts:
        words.extend(text.split())

    # Hitung frekuensi untuk N-gram
    if n == 1:  # Unigram
        freq = Counter(words)
        title = f'Top {top} Kata dalam Ulasan {sentiment.capitalize()}'
        filename = f'top_words_{sentiment}.png'
    else:  # N-gram
        n_grams = []
        for text in texts:
            tokens = text.split()
            n_grams.extend(list(ngrams(tokens, n)))
        freq = Counter(n_grams)
        title = f'Top {top} {n}-gram dalam Ulasan {sentiment.capitalize()}'
        filename = f'top_{n}gram_{sentiment}.png'

    # Plot
    plt.figure(figsize=(12, 8))
    common_words = dict(freq.most_common(top))

    if n == 1:
        sns.barplot(x=list(common_words.values()), y=list(common_words.keys()))
    else:
        labels = [' '.join(ng) for ng in common_words.keys()]
        sns.barplot(x=list(common_words.values()), y=labels)

    plt.title(title)
    plt.xlabel('Frekuensi')
    plt.tight_layout()
    plt.show()

# Plot top unigrams untuk setiap sentimen
for sentiment in ['negative', 'neutral', 'positive']:
    plot_top_ngrams(df, sentiment, n=1, top=20)

# Plot top bigrams untuk setiap sentimen
for sentiment in ['negative', 'neutral', 'positive']:
    plot_top_ngrams(df, sentiment, n=2, top=20)
```

## **6. Analisis Sentimen dengan Metode Machine Learning**

### **6.1 Persiapan Data**

```python
# Memisahkan data dan label
X = df['processed_text']
y = df['sentiment']

# Encoding label
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Simpan encoder untuk digunakan nanti
with open('label_encoder.pkl', 'wb') as f:
    pickle.dump(label_encoder, f)
```

### **6.2 Percobaan 1: SVM dengan TF-IDF (80/20)**

#### **6.2.1 Persiapan Data dan Ekstraksi Fitur**

```python
# Split data 80/20
X_train_1, X_test_1, y_train_1, y_test_1 = train_test_split(X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded)

# TF-IDF Vectorizer
tfidf_vectorizer = TfidfVectorizer(max_features=5000)
X_train_tfidf_1 = tfidf_vectorizer.fit_transform(X_train_1)
X_test_tfidf_1 = tfidf_vectorizer.transform(X_test_1)

# Simpan vectorizer untuk digunakan nanti
with open('tfidf_vectorizer.pkl', 'wb') as f:
    pickle.dump(tfidf_vectorizer, f)
```

#### **6.2.2 Training Model SVM**

```python
print("\nMemulai pelatihan model SVM dengan TF-IDF...")
start_time = time.time()

# Training model SVM
svm_model = SVC(kernel='linear', probability=True, random_state=42)
svm_model.fit(X_train_tfidf_1, y_train_1)

training_time = time.time() - start_time
print(f"Waktu pelatihan: {training_time:.2f} detik")

# Simpan model
with open('svm_tfidf_model.pkl', 'wb') as f:
    pickle.dump(svm_model, f)
```

#### **6.2.3 Evaluasi Model SVM**

```python
# Prediksi pada data testing
y_pred_svm = svm_model.predict(X_test_tfidf_1)

# Evaluasi model
print("\nHasil Evaluasi Model SVM dengan TF-IDF (80/20):")
print(classification_report(y_test_1, y_pred_svm, target_names=label_encoder.classes_))

# Plot confusion matrix
plt.figure(figsize=(8, 6))
cm = confusion_matrix(y_test_1, y_pred_svm)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
plt.title('Confusion Matrix - SVM dengan TF-IDF')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

### **6.3 Percobaan 2: Random Forest dengan Word2Vec (80/20)**

#### **6.3.1 Persiapan Data dan Ekstraksi Fitur dengan Word2Vec**

```python
# Split data 80/20 (menggunakan split yang sama dengan percobaan 1)
X_train_2, X_test_2, y_train_2, y_test_2 = X_train_1, X_test_1, y_train_1, y_test_1

# Tokenisasi untuk Word2Vec
tokenized_train = [text.split() for text in X_train_2]

# Melatih model Word2Vec
print("\nMemulai pelatihan model Word2Vec...")
w2v_model = Word2Vec(sentences=tokenized_train, vector_size=100, window=5, min_count=1, workers=4, sg=1)

# Simpan model Word2Vec
w2v_model.save('word2vec_model.model')

# Fungsi untuk mengubah teks menjadi vektor dengan Word2Vec
def text_to_vector(text, model):
    """
    Mengubah teks menjadi vektor dengan rata-rata vektor kata dari Word2Vec
    """
    words = text.split()
    vectors = [model.wv[word] for word in words if word in model.wv]
    if len(vectors) == 0:
        return np.zeros(model.vector_size)
    return np.mean(vectors, axis=0)

# Mengubah dokumen teks menjadi vektor
X_train_w2v = np.array([text_to_vector(text, w2v_model) for text in X_train_2])
X_test_w2v = np.array([text_to_vector(text, w2v_model) for text in X_test_2])
```

#### **6.3.2 Training Model Random Forest dengan Word2Vec**

```python
print("\nMemulai pelatihan model Random Forest dengan Word2Vec...")
start_time = time.time()

# Training model Random Forest
rf_w2v_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_w2v_model.fit(X_train_w2v, y_train_2)

training_time = time.time() - start_time
print(f"Waktu pelatihan: {training_time:.2f} detik")

# Simpan model
with open('rf_w2v_model.pkl', 'wb') as f:
    pickle.dump(rf_w2v_model, f)
```

#### **6.3.3 Evaluasi Model Random Forest dengan Word2Vec**

```python
# Prediksi pada data testing
y_pred_rf_w2v = rf_w2v_model.predict(X_test_w2v)

# Evaluasi model
print("\nHasil Evaluasi Model Random Forest dengan Word2Vec (80/20):")
print(classification_report(y_test_2, y_pred_rf_w2v, target_names=label_encoder.classes_))

# Plot confusion matrix
plt.figure(figsize=(8, 6))
cm = confusion_matrix(y_test_2, y_pred_rf_w2v)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
plt.title('Confusion Matrix - Random Forest dengan Word2Vec')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

## **7. Analisis Sentimen dengan Deep Learning**

### **7.1 Persiapan Data untuk Deep Learning**

```python
# Menyiapkan tokenizer Keras
max_features = 10000  # Jumlah kata dalam vocabulary
maxlen = 200  # Panjang maksimal sequence

tokenizer = Tokenizer(num_words=max_features)
tokenizer.fit_on_texts(X_train_1)

# Mengubah teks menjadi sequence
X_train_seq = tokenizer.texts_to_sequences(X_train_1)
X_test_seq = tokenizer.texts_to_sequences(X_test_1)

# Padding sequence
X_train_pad = pad_sequences(X_train_seq, maxlen=maxlen)
X_test_pad = pad_sequences(X_test_seq, maxlen=maxlen)

# Simpan tokenizer
with open('tokenizer.pkl', 'wb') as f:
    pickle.dump(tokenizer, f)
```

### **7.2 Membangun Model LSTM**

#### **7.2.1 Definisi Arsitektur Model**

```python
# Definisi model
model = Sequential()
model.add(Embedding(max_features, 128, input_length=maxlen))
model.add(Bidirectional(LSTM(64, dropout=0.2, recurrent_dropout=0.2)))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(3, activation='softmax'))  # 3 kelas: negative, neutral, positive

# Kompilasi model
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

#### **7.2.2 Training Model LSTM**

```python
# Callback untuk early stopping
early_stopping = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)

# Simpan detail model
model.summary()

print("\nMemulai pelatihan model LSTM...")
start_time = time.time()

# Training model
history = model.fit(
    X_train_pad, y_train_1,
    epochs=10,
    batch_size=32,
    validation_split=0.1,
    callbacks=[early_stopping]
)

training_time = time.time() - start_time
print(f"Waktu pelatihan: {training_time:.2f} detik")

# Simpan model
model.save('lstm_model.h5')
```

#### **7.2.3 Evaluasi Model LSTM**

```python
# Visualisasi training & validation loss
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Training & Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
```

```python
# Visualisasi training & validation accuracy
plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Training & Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.tight_layout()
plt.show()
```
